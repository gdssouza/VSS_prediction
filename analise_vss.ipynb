```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("vss_xyz_dia_2017-2019",index_col=0)
df.DATE = pd.to_datetime(df.DATE)
df.set_index('DATE',inplace=True)
df
```

```python
#visualizando
plt.figure(figsize=(18,9))

plt.subplot(4,1,1)
df.VSSX.plot()
plt.ylabel("X (nT)")

plt.subplot(4,1,2)
df.VSSY.plot()
plt.ylabel("Y (nT)")

plt.subplot(4,1,3)
df.VSSZ.plot()
plt.ylabel("Z (nT)")

plt.subplot(4,1,4)
df.F.plot()
plt.ylabel("F (nT)")

plt.show()
```

### Analisando tendÃªncias

```python
periodo = 21
#visualizando
plt.figure(figsize=(18,9))

plt.subplot(4,1,1)
df.VSSX.rolling(periodo).mean().plot()
plt.ylabel("X (nT)")

plt.subplot(4,1,2)
df.VSSY.rolling(periodo).mean().plot()
plt.ylabel("Y (nT)")

plt.subplot(4,1,3)
df.VSSZ.rolling(periodo).mean().plot()
plt.ylabel("Z (nT)")

plt.subplot(4,1,4)
df.F.rolling(periodo).mean().plot()
plt.ylabel("F (nT)")

plt.show()
```

### Analisando sazonalidade

```python
#visualizando
plt.figure(figsize=(18,9))

plt.subplot(4,1,1)
df.VSSX.diff(1).plot()
plt.ylabel("X (nT)")

plt.subplot(4,1,2)
df.VSSY.diff(1).plot()
plt.ylabel("Y (nT)")

plt.subplot(4,1,3)
df.VSSZ.diff(1).plot()
plt.ylabel("Z (nT)")

plt.subplot(4,1,4)
df.F.diff(1).plot()
plt.ylabel("F (nT)")

plt.show()
```
